```python
# Question 1: Write a Python program to reverse a string without using any built-in string reversal functions.

listNum = [2,3,4,67,1,6,8,12,15]
print(f"List before reversal: {listNum}")
i=0
j=len(listNum)-1
while i<=j:
    temp = listNum[i]
    listNum[i] = listNum[j]
    listNum[j] = temp
    i+=1
    j-=1
print(f" List after reversal: {listNum}")
```

```python
#Question 2. Implement a function to check if a given string is a palindrome.
palinStr = "abcdedcba"
print(f"String to check for palindrome: {palinStr}")
i=0
j=len(palinStr)-1
isPalindrome = True
while i<=j:
    if palinStr[i] != palinStr[j]:
        isPalindrome = False
        break
    i+=1
    j-=1
print(f"String {palinStr} is palindrome: {isPalindrome}")
```

```python
# Question 3. Write a program to find the largest element in a given list.
listNum = [2,3,4,67,1,6,8,12,15]
print(f"Largest element from list: {listNum}")
i=0
maxNum=0
while i<len(listNum):
    if maxNum<listNum[i]:
        maxNum=listNum[i]
    i+=1

print(f"Largest element is: {maxNum}")
```

```python
# Question 4. Implement a function to count the occurrence of each element in a list.
listNum = [1,1,2,3,2,1,2,3,2,1]
print(f"Count the occurrence of each element from list: {listNum}")
mydict = {}
for num in listNum:
    if num in mydict:
        temp = mydict.get(num)
        mydict[num] = temp+1
    else:
        mydict[num] = 1
print(f"Below are details")        
for key in mydict:
    print(f"Element:{key}, Occurance:{mydict[key]}")
```

```python
# Question 5. Write a Python program to find the second largest number in a list.
listNum = [2,3,4,67,32,66,45,12,15]
print(f"Second largest element from list: {listNum}")
i=0
maxNum=0
while i<len(listNum):
    if maxNum<listNum[i]:
        maxNum=listNum[i]
    i+=1

i=0
secondMaxNum=0    
while i<len(listNum):
    if secondMaxNum<listNum[i]<maxNum:
        secondMaxNum=listNum[i]
    i+=1
print(f"2nd largest element: {secondMaxNum}")
```

```python
# Question 6. Implement a function to remove duplicate elements from a list.
listNum = [1,1,2,3,2,1,2,3,2,1,4,5,4]
print(f"Remove duplicate elements from a list: {listNum}")
mydict = {}
i=0
while i<len(listNum):
    if listNum[i] in mydict:
        listNum.pop(i)
        i-=1
    else:
        mydict[listNum[i]] = 1
    i+=1
print(f"List with distinct elements: {listNum}")
```

```python
# Question 7. Write a program to calculate the factorial of a given number.
num = 5
print(f"Find the factorial of number:{num}")
factorial = 1
for i in range(1,num+1):
    factorial = factorial*i
print(f"Factorial of number:{num} is {factorial}")
```

```python
# Question 8. Implement a function to check if a given number is prime.
# Prime number is a number which has ONLY and ONLY 2 factors.
# Hence that would be 1 and the number itself.
num =18
print(f"Check number {num} is Prime?")
countFactor=0
if num>1:
    countFactor=1
    for i in range(2, (num//2)+1):
        if num%i==0:
            countFactor+=1
        if countFactor>=2:
            break
if countFactor==1:
    print(f"Number {num} is Prime")
else:
    print(f"Number {num} is Not Prime")
```

```python
# Question 9. Write a Python program to sort a list of integers in ascending order.
from itertools import combinations

listNum = [2,3,4,67,32,66,45,12,15]
print(f"Input list : {listNum}")

for i,j in combinations(range(len(listNum)), r=2):
    if listNum[i]>listNum[j]:
        temp = listNum[i]
        listNum[i]=listNum[j]
        listNum[j]=temp
print(f"Ascending list : {listNum}")
```

```python
# Question 10. Implement a function to find the sum of all numbers in a list.
#Input list of numbers
listNum = [2,3,4,67,32,66,45,12,15]
print(f"Input list : {listNum}")

# First Approach
print(f"Summ of all numbers in a list : {sum(listNum)}") 

# Second Approach
sumNum=0
for num in listNum:
    sumNum+=num
print(f"Summ of all numbers in a list : {sumNum}")
```
