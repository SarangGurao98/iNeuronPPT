```python
#11. Write a program to find the common elements between two lists.
list1 = [1, 2, 4, 5, 6,7, 11]
list2 = [1, 3, 4, 6,9,10, 11]

print(f"First list  :{list1}")
print(f"Second list :{list2}")
commonList=[]
for num in list1:
    if num in list2:
        commonList.append(num)
print(f"Common list :{commonList}")
```

```python
#12. Implement a function to check if a given string is an anagram of another string.
# Sample inputs
s,t = "Astronomer", "Moon starer"
#s,t = "a gentleman","elegant man"
#s,t = "Fourth of July","joyful Fourth"

print(f"Checking for anagram")
print(f"Two string are :{s}: & :{t}:")
# Removing all spaces from string
s = s.replace(" ","")
t = t.replace(" ","")
# All string to lowercase
s = s.lower()
t = t.lower()
isAnagram = False # bool parameter for anagram
if len(s)==len(t):
    for i in range(len(s)):
        t = t.replace(s[i],'',1) # Replacing character by character
if len(t)==0:
    isAnagram = True
print(f"Is anagram :{isAnagram}") 
```

```python
#13. Write a Python program to generate all permutations of a given string.
from itertools import permutations

str = "abcd"
# Input string
print(f"Input string :{str}")
 
# Finding all permutation
perm = permutations(str)

permList = []
for tempStr in perm:
    x = "".join(tempStr)
    permList.append(x)

# Output all permutations of a given string.
print("All permutations of a given input string:")
print(f"{permList}")
```

```python
#14. Implement a function to calculate the Fibonacci sequence up to a given number of terms.

fibNumTerm = 11
# Input - Fibonacci series 
print(f"Display Fibonacci total number terms: {fibNumTerm}")

fibSeries = []
count = 0
while count<fibNumTerm:
    if len(fibSeries)==0:
        i=0
    else:
        if i==0:
            i=1
        else:
            i = fibSeries[-1]+fibSeries[-2]
    fibSeries.append(i)
    count+=1
print(f"Fibonacci sequence {fibNumTerm} numbers:")
print(f"{fibSeries}")
```

```python
#15. Write a program to find the median of a list of numbers.

listNum = [1, 2, 3, 4, 5, 6, 7]    # Odd Sample
#listNum = [1, 2, 3, 4, 5, 6, 7,8] # Even Sample
#listNum = [1, 3, 3, 6, 7, 8, 9]    # Odd Sample
#listNum = [1, 2, 3, 4, 5, 6, 8, 9] # Even Sample

# Input List of which we have to find median
print(f"Input list :{listNum}")

# We have to check if length of list is odd or even as values to calculate median will change accordingly
medianValue = 0.0
mid = 0
if (len(listNum)%2)==1:
    mid = (len(listNum)//2)
    medianValue = listNum[mid]
else:
    mid = len(listNum)//2
    medianValue = listNum[mid-1]+listNum[mid]
    medianValue = medianValue/2
print(f"The median of a list of numbers :{medianValue}")
```

```python
#16. Implement a function to check if a given list is sorted in non-decreasing order.

ascendingList = [1.2,3,4,3,6,7] # False Sample
#ascendingList = [1,2,3,4,5,6,7] # True Sample
print(f"Check non-decreasing order - Input List :{ascendingList}")

isAscending = True
i=1
if len(ascendingList)>1:
    while isAscending and i<len(ascendingList):
        if ascendingList[i-1]>ascendingList[i]:
            isAscending = False
            break
        i+=1
print(f"List is sorted in non-decreasing order? {isAscending}")
```

```python
#17. Write a Python program to find the intersection of two lists.

# Initialize two list : ListA and ListB
listA = [1,2,3,4,9,6,7] 
listB = [1,5,3,8,6,7,2]
print("Find the intersection of below two lists.")
print(f"List #1:{listA} , List #2:{listB}")

# Initialize Output list 
intersectionList = []
# Iterating through elment of ListA and checking those element if ListB.
# if present add that element to outputList => intersectionList
for a in listA:
    if a in listB:
        intersectionList.append(a)
if len(intersectionList)!=0:
    print(f"Intersection elements are :{intersectionList}")
else:
    print(f"Unique Lists: No Intersection elements")
```

```python
#18. Implement a function to find the maximum subarray sum in a given list.
from itertools import combinations

arrNum = [-2, -3, 4, -1, -2, 1, 5, -3]       # Output: [4, -1, -2, 1, 5]  # Sample 1
#Input initialization
print(f"find the maximum subarray sum in a given list : {arrNum}")

maxSum = 0       #Initializing parament for check max of sum.
maxSubArray = [] #Initializing list which has max of sum.
# checking for all possible subarrays combinations and checking sums and fetching maximum of it.
for i,j in combinations(range(len(arrNum)+1), r=2):
    tempSum = sum(arrNum[i:j])
    if tempSum > maxSum:
        maxSum = tempSum
        maxSubArray = arrNum[i:j]
        
print(f"Maximum subarray : {maxSubArray}")
```

```python
#19. Write a program to remove all vowels from a given string.

#Initializing input String from where vowels character need to be removed.
inputStr = "SarangAEjhdfgIOjklhsakljfhUlk"
print(f"Input String :{inputStr}")

# Initializing VOWELS string for checking purpose
vowelStr = ['A','E','I','O','U']
# Iterating through input String and checking for vowels. 
# If found - then remove it by using String replace method
for charS in inputStr:
    if charS.upper() in vowelStr:
        inputStr=inputStr.replace(charS,"")
print(f"Output String :{inputStr}")
```

```python
#20. Implement a function to reverse the order of words in a given sentence.

inputStr = "Hello World, I am from India"
inputStr  = inputStr.strip()
print(f"Input Srting :{inputStr}:")

revInputStr =""
temp=""
for i in range(len(inputStr)-1,-1,-1):
    if ord(inputStr[i])!=32:
        temp=inputStr[i]+temp
    else:
        temp= temp+" "
        revInputStr+=temp
        temp=""
if len(temp)>0:
    revInputStr+=temp
print(f"Output Srting :{revInputStr}")
```
