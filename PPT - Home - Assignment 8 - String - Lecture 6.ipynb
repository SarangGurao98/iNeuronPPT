### Question 1

```python
"""
Question 1

Given two strings s1 and s2, return *the lowest **ASCII** sum of deleted characters to make two strings equal*.

Example 1:
Input: 
    s1 = "sea", s2 = "eat"
Output: 231
Explanation: Deleting "s" from "sea" adds the ASCII value of "s" (115) to the sum.
             Deleting "t" from "eat" adds 116 to the sum.
             At the end, both strings are equal, and 115 + 116 = 231 is the minimum sum possible to achieve this.
"""
# Input Sample 1 for s1, s2
str_s1, str_s2 = "sea","eat"
#str_s1, str_s2 = "delete","leet"

print(f"String s1:{str_s1}, s2:{str_s2}")

#initializing dp woth row = len(s1)+1 and column = len(s2)+1
dp = [[0 for j in range(len(str_s2)+1)] for i in range(len(str_s1)+1)] 

# Traversing in reverse way from max to 0
for i in range(len(dp)-1,-1,-1):
    for j in range(len(dp[0])-1,-1,-1):
        if i==len(dp)-1 and j==len(dp[0])-1:
            dp[i][j]=0  # Initializing last element [len][len]==0
        elif i==len(dp)-1:
            dp[i][j] = ord(str_s2[j])+dp[i][j+1] # Initializing ASCII Value for s2
        elif j==len(dp[0])-1:
            dp[i][j] = ord(str_s1[i])+dp[i+1][j]  # Initializing ASCII Value for s1
        else:
            if ord(str_s1[i]) == ord(str_s2[j]):  # if both character are same
                dp[i][j] = dp[i+1][j+1]
            else:                                 # Checking for least cost path for string
                dp[i][j] = min(ord(str_s1[i]) + dp[i+1][j], ord(str_s2[j])+dp[i][j+1])
        
print(f"Output :{dp[0][0]}")   
```

### Question 2 

```python
"""
Question 2
Given a string s containing only three types of characters: '(', ')' and '*', return true *if* s *is **valid***.
The following rules define a **valid** string:
- Any left parenthesis '(' must have a corresponding right parenthesis ')'.
- Any right parenthesis ')' must have a corresponding left parenthesis '('.
- Left parenthesis '(' must go before the corresponding right parenthesis ')'.
- '*' could be treated as a single right parenthesis ')' or a single left parenthesis '(' or an empty string "".

Example 1:  Input: s = "()"   Output: true
Example 2:  Input: s = "(*)"  Output: true
Example 3:  Input: s = "(*))" Output: true
"""
s = "(*))"
print(f"Input String:{s}")

openBraces,closeBraces,valid_s = 0,0,True
slen= len(s)-1
for i in range(len(s)):
    if s[i] == '(' or s[i] == '*':
        openBraces+=1
    else: 
        openBraces-=1
    if s[slen-i] == ')' or s[slen-i] == '*':
        closeBraces+=1
    else:
        closeBraces-=1
    if openBraces < 0 or closeBraces<0:
        valid_s= False
print(f"Valid String?{valid_s}")
```

### Question 3 

```python
"""
Question 3
Given two strings word1 and word2, return *the minimum number of **steps** required to make* word1 *and* word2 *the same*.
In one **step**, you can delete exactly one character in either string.

Example 1:
word1 = "sea", word2 = "eat"
Output: 2
Explanation: You need one step to make "sea" to "ea" and another step to make "eat" to "ea".

"""
word1, word2 = "sea","eat"
#word1, word2 = "leetcode", "etco"
#word1, word2 = "park", "spake" #output = 3
#word1, word2 = "sea", "ate" #output = 4
#word1, word2 = "intention", "execution" #output = 8
#word1, word2 = "dinitrophenylhydrazine", "acetylphenylhydrazine"
print(f"Given two string are : {word1}, {word2}")

m, n = len(word1), len(word2)
# Create a 2D array to store the subproblem solutions
dp = [[0] * (n + 1) for _ in range(m + 1)]
# Base case: if either string is empty, the answer is the length of the other string
for i in range(m + 1):
    dp[i][0] = i
for j in range(n + 1):
    dp[0][j] = j
# Dynamic programming to fill up the DP table
for i in range(1, m + 1):
    for j in range(1, n + 1):
        if word1[i - 1] == word2[j - 1]:
            dp[i][j] = dp[i - 1][j - 1]
        else:
            dp[i][j] = 1 + min(dp[i - 1][j], dp[i][j - 1])
# The last cell in the DP table contains the answer
print(f"Total Step :{dp[m][n]}")
```

```python
"""
Question 4
You need to construct a binary tree from a string consisting of parenthesis and integers.
The whole input represents a binary tree. It contains an integer followed by zero, one or two pairs of parenthesis. 
The integer represents the root's value and a pair of parenthesis contains a child binary tree with the same structure.
You always start to construct the **left** child node of the parent first if it exists.

Input:s = "4(2(3)(1))(6(5))"
Output:[4,2,6,3,1,5]

"""
class Node:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
def node2List(root):
    # As we need output in level sequence. 
    listTree = []    #output
    if root is None:
        return
    q = []
    q.append(root)
    while q:
        count = len(q)
        while count > 0:
            temp = q.pop(0)
            listTree.append(temp.val)
            if temp.left:
                q.append(temp.left)
            if temp.right:
                q.append(temp.right)
            count -= 1
    return listTree

# Input s in a particular format
s = "4(2(3)(1))(6(5))"
outList = []                  # output list     
root = Node(int(s[0]))        # first element will be root
listTN = []                   # stack for tree
for i in range(1,len(s)):     # Binary Tree
    if (s[i] == '('):
        listTN.append(root)   
    elif (s[i] == ')'):
        root = listTN[-1]
        del listTN[-1]
    else:
        if (root.left == None):
            left = Node(int(s[i]))
            root.left = left
            root = root.left
        elif (root.right == None):
            right = Node(int(s[i]))
            root.right = right
            root = root.right
#passing root to method 
outList = node2List(root)
print(outList)
```
