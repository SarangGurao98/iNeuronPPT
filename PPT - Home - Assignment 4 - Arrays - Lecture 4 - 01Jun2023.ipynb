```python
"""
Question 1 
Given three integer arrays arr1, arr2 and arr3 sorted in strictly increasing order, 
return a sorted array of only the integers that appeared in all three arrays.

Example 1:
Input: arr1 = [1,2,3,4,5], arr2 = [1,2,5,7,9], arr3 = [1,3,4,5,8]
Output: [1,5]

Explanation: Only 1 and 5 appeared in the three arrays.
"""

arr1 = [1,2,3,4,5] #Input Sorted Array 1
arr2 = [1,2,5,7,9] #Input Sorted Array 2
arr3 = [1,3,4,5,8] #Input Sorted Array 3
print("Input array 1:",arr1)
print("Input array 2:",arr2)
print("Input array 3:",arr3)

#Output = [1,5]
Output_arr = [] # Param for storing common element
Output_arr_index = 0 # Index for common element to enter
temp_arr = [] # Adding all three arrays

# Merged all arrays in one single array
temp_arr+=arr1
temp_arr+=arr2
temp_arr+=arr3
# Sort the Merged array
temp_arr.sort()
#print("Merged array ",temp_arr)

#index for iterating through merged array
i=0
while i<len(temp_arr):
    if ((i+2)<len(temp_arr)): #to compare i+2 is not array out of index
         if temp_arr[i] == temp_arr[i+1] and temp_arr[i] == temp_arr[i+2]: #Check three elements from merged arrays
            Output_arr.insert(Output_arr_index, temp_arr[i]) #inserting the common element
            Output_arr_index+=1 #increment index count of common element array
            i+=2 # As we compare 3 element, counter increment next 2
    i+=1 #counter increment by default to iterate throughh merged array    

print("Common Elements from Arrays :",Output_arr)
```

### Question 4 

```python
"""
Question 4 
Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2), ..., 
(an, bn) such that the sum of min(ai, bi) for all i is maximized. Return the maximized sum.

Example 1:
Input: nums = [1,4,3,2]
Output: 4

Explanation: All possible pairings (ignoring the ordering of elements) are:
1.	(1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3
2.	(1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3
3.	(1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4
So the maximum possible sum is 4.

"""
import itertools

# Integer array of 2n integers
arr_nums = [1,4,3,2]

arr_nums.sort()
subsets_arr = list(itertools.combinations(arr_nums, 2))  
print("Sub Array for 2n array",subsets_arr )

i=0;
j=len(subsets_arr)
mid = int((i+j)/2)

j=j-1
arr_nums_comb=[]
while i<mid:
    #print(f"i:{i}, j:{j}, mid:{mid}")
    arr_num_temp=[]
    arr_num_temp.append(subsets_arr[i])
    arr_num_temp.append(subsets_arr[j])
    arr_nums_comb.append(arr_num_temp)
    i=i+1
    j=j-1

print("Sub Arrays :",arr_nums_comb )

# Finding the minimal of element
maxOutput = 0
for i in arr_nums_comb:
    temp1 = min(i[0])
    temp2 = min(i[1])
    currentOutput = temp1 + temp2
    if maxOutput < currentOutput:
         maxOutput = currentOutput


print(f"Maximum possible sum :{maxOutput}")
```

```python
"""
Question 6 
Given an integer array nums sorted in non-decreasing order, 
return an array of the squares of each number sorted in non-decreasing order.

Example 1:
Input: nums = [-4,-1,0,3,10]
Output: [0,1,9,16,100]

Explanation: After squaring, the array becomes [16,1,0,9,100]. After sorting, it becomes [0,1,9,16,100]

"""
input_nums = [-4,-1,0,3,10]
#Expected Output: [0,1,9,16,100]
print("input_nums",input_nums)

for i in range(len(input_nums)):
    input_nums[i] =  input_nums[i] * input_nums[i]
input_nums.sort()    
print("input_nums",input_nums)
```

```python
"""
Question 7 
You are given an m x n matrix M initialized with all 0's and an array of operations ops, 
where ops[i] = [ai, bi] means M[x][y] should be incremented by one for all 0 <= x < ai and 0 <= y < bi.
Count and return the number of maximum integers in the matrix after performing all the operations

Example 1:

Refer IMAGE in file Assignment 4 Questions - Arrays _ DSA

Input: m = 3, n = 3, ops = [[2,2],[3,3]]
Output: 4
Explanation: The maximum integer in M is 2, and there are four of it in M. So return 4.

"""
m=3 #Rows 
n=3 #Cols
m_matrix = [[0 for i in range(n)] for j in range(m)] #Initialize array for m*n matrix 
ops = [[2,2],[3,3]] # Ops array sample input 1
#ops = [[2,2],[3,3],[3,3],[3,3],[2,2],[3,3],[3,3],[3,3],[2,2],[3,3],[3,3],[3,3]]  # Ops array sample input 2
#ops = []  # Ops array sample input 3

print("M[x][y]:", m_matrix)
print("ops[i] = [ai, bi] :", ops)

# Iterating and increamenting according with ops elements
for tempOps in ops:
    tempRow = tempOps[0]
    tempCol = tempOps[1]
    for i in range(tempRow):
        for j in range(tempCol):
            m_matrix[i][j] = m_matrix[i][j]+1

# find maximum integer occurance from change matrix
maxNumber = 0
maxCount = 0
temp = 0
for i in range(m):
    for j in range(n):
        temp = m_matrix[i][j]
        if temp>=maxNumber:
            maxNumber = temp;
            maxCount+=1    
# Output
print("M[x][y]:", m_matrix)
print("maxNumber:", maxNumber)
print("maxCount:", maxCount)

```
